```python
import numpy as np

from dance.modules.single_modality.cell_type_annotation.actinn import ACTINN
from dance.utils import set_seed
def getAuc_Actinn(tissue,train_dataset,test_dataset,hidden_dims=[256, 256], lambd=0.01, device='cpu',lr=0.001, num_epochs=21,
              batch_size=1000, print_cost=True):
    # Initialize model and get model specific preprocessing pipeline
    model = ACTINN(hidden_dims=hidden_dims, lambd=lambd, device=device)
    preprocessing_pipeline = model.preprocessing_pipeline(normalize=True, filter_genes=True)

    # Load data and perform necessary preprocessing
    dataset = ScDeepSortDataset(species="human", tissue=tissue,
                                train_dataset=train_dataset, test_dataset=test_dataset)
    data = dataset.load_data(transform=preprocessing_pipeline, cache=False)
    # Obtain training and testing data
    x_train, y_train = data.get_train_data(return_type="torch")
    x_test, y_test = data.get_test_data(return_type="torch")
    # Train and evaluate model
    set_seed(42)
    model.fit(x_train, y_train, lr=lr, num_epochs=num_epochs,
              batch_size=batch_size, print_cost=print_cost)
    score=model.score(x_test, y_test)
    print(f"ACC: {score:.4f}")
    print(y_test.shape)
    return score
```

```python
Actinn_scores=(getAuc_Actinn("CD8",["1641","1357"],["492"]),getAuc_Actinn("CD4",["1247"],["559"]),getAuc_Actinn("Immune",["11407"],["7572"]),getAuc_Actinn("Spleen",["3043","4657","4362","4115"],["1729","2125","2184"]),getAuc_Actinn("Brain",["328"],["138"]))
```

```python
import numpy as np
import torch
from dance.modules.single_modality.cell_type_annotation.scdeepsort import ScDeepSort
from dance.utils import set_seed
def getAuc_ScDeepSort(tissue,train_dataset,test_dataset,batch_size=500,cache=False,dense_dim=400,hidden_dim=200,n_layers=1,dropout=0.1,device="cpu",n_epochs=300,lr=1e-3
                        ,weight_decay=5e-4,test_rate=0.2):
     # Initialize model and get model specific preprocessing pipeline
    model = ScDeepSort(dense_dim, hidden_dim, n_layers, "human", tissue, dropout=dropout,
                       batch_size=batch_size, device=device)
    preprocessing_pipeline = model.preprocessing_pipeline(n_components=dense_dim)

    # Load data and perform necessary preprocessing
    dataloader = ScDeepSortDataset(species="human", tissue=tissue, test_dataset=test_dataset,
                                   train_dataset=train_dataset)
    data = dataloader.load_data(transform=preprocessing_pipeline, cache=cache)

    # Obtain training and testing data
    y_train = data.get_y(split_name="train", return_type="torch").argmax(1)
    y_test = data.get_y(split_name="test", return_type="torch")
    num_labels = y_test.shape[1]

    # Get cell feature graph for scDeepSort
    # TODO: make api for the following block?
    g = data.data.uns["CellFeatureGraph"]
    num_genes = data.shape[1]
    gene_ids = torch.arange(num_genes)
    train_cell_ids = torch.LongTensor(data.train_idx) + num_genes
    test_cell_ids = torch.LongTensor(data.test_idx) + num_genes
    g_train = g.subgraph(torch.concat((gene_ids, train_cell_ids)))
    g_test = g.subgraph(torch.concat((gene_ids, test_cell_ids)))

    # Train and evaluate the model
    model.fit(g_train, y_train, epochs=n_epochs, lr=lr, weight_decay=weight_decay,
              val_ratio=test_rate)
    score = model.score(g_test, y_test)
    print(f"{score=:.4f}")
    return score
```

```python
ScDeepSort_scores=(getAuc_ScDeepSort("CD8",["1641","1357"],["492"]),getAuc_ScDeepSort("CD4",["1247"],["559"]),getAuc_ScDeepSort("Immune",["11407"],["7572"]),getAuc_ScDeepSort("Spleen",["3043","4657","4362","4115"],["1729","2125","2184"]),getAuc_ScDeepSort("Brain",["328"],["138"],dense_dim=100))
```

```python
import numpy as np
import torch
from dance.modules.single_modality.cell_type_annotation.singlecellnet import SingleCellNet
from dance.utils import set_seed
def getAuc_SingleCellNet(tissue,train_dataset,test_dataset,num_trees=1000,normalize=False,num_top_genes=100,num_top_gene_pairs=250,cache=False,stratify=True,num_rand=100,random_state=10):
    # Initialize model and get model specific preprocessing pipeline
    model = SingleCellNet(num_trees=num_trees)
    preprocessing_pipeline = model.preprocessing_pipeline(normalize=normalize, num_top_genes=num_top_genes,
                                                          num_top_gene_pairs=num_top_gene_pairs)

    # Load data and perform necessary preprocessing
    dataloader = ScDeepSortDataset(train_dataset=train_dataset, test_dataset=test_dataset,
                                   species="human", tissue=tissue)
    data = dataloader.load_data(transform=preprocessing_pipeline, cache=cache)

    # Obtain training and testing data
    x_train, y_train = data.get_train_data(return_type="numpy")
    y_train = y_train.argmax(1)  # convert one-hot representation into label index representation
    x_test, y_test = data.get_test_data(return_type="numpy")
    print(y_train.shape)
    # Train and evaluate the model
    model.fit(x_train, y_train, stratify=stratify, num_rand=num_rand, random_state=random_state)
    y_test = y_test.argmax(1)
    # print(y_test.shape)
    score = model.score(x_test, y_test,score_func="ari")
    print(f"{score=:.4f}")
    return score
```

```python
SingleCellNet_scores=(getAuc_SingleCellNet("CD8",["1641","1357"],["492"],cache=True,normalize=True),getAuc_SingleCellNet("CD4",["1247"],["559"],normalize=True),getAuc_SingleCellNet("Immune",["11407"],["7572"],normalize=True),getAuc_SingleCellNet("Spleen",["3043","4657","4362","4115"],["1729","2125","2184"],normalize=True),getAuc_SingleCellNet("Brain",["328"],["138"],normalize=True)) #好像只能用normalize,然后用ari
```

```python
import numpy as np
import torch
from dance.modules.single_modality.cell_type_annotation.svm import SVM
from dance.utils import set_seed
import argparse
def getAuc_SVM(tissue,train_dataset,test_dataset,dense_dim=400,log_level="INFO",cache=False):
    # Initialize model and get model specific preprocessing pipeline
    model = SVM(argparse.Namespace(**{"species":"human","tissue":tissue}))  # TODO: get useful args out
    preprocessing_pipeline = model.preprocessing_pipeline(n_components=dense_dim, log_level=log_level)

    # Load data and perform necessary preprocessing
    dataloader = ScDeepSortDataset(train_dataset=train_dataset, test_dataset=test_dataset,
                                   species="human", tissue=tissue)
    data = dataloader.load_data(transform=preprocessing_pipeline, cache=cache)

    # Obtain training and testing data
    x_train, y_train = data.get_train_data()
    y_train_converted = y_train.argmax(1)  # convert one-hot representation into label index representation
    x_test, y_test = data.get_test_data()

    # Train and evaluate the model
    model.fit(x_train, y_train_converted)
    score = model.score(x_test, y_test)
    print(f"{score=:.4f}")
    return score
```
